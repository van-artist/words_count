```python
target_words_str="taxes resident revenue Transportation employment destruction cruise industry revenue revenue Tourist numbers cost government intervention"
target_words=target_words_str.split(" ")
# 大小写不区分，添加大写和小写形式,还有首字母大写形式
target_words_lower=[x.lower() for x in target_words]
target_words_upper=[x.upper() for x in target_words]
target_words_first_upper=[x.capitalize() for x in target_words]
target_words=target_words+target_words_lower+target_words_upper+target_words_first_upper
```

```python
print("target_words: ", target_words)
```
